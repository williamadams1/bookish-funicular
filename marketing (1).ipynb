```python
import pandas as pd 
import numpy as np
import seaborn as sns 
%matplotlib inline 
import matplotlib.pyplot as plt 
import sklearn 
from sklearn import preprocessing

marketing = pd.read_csv("marketingdataset.csv")
marketing.head()
```

```python
col_names = ['cust_id', 'gender', 'age', 'marital_status', 'country', 'income', 'education', 'job', 'adoption']
marketing.columns = col_names
marketing.head()
```

```python
plt.scatter(x = "marital_status", y = "age", data = marketing)
```

```python
sns.factorplot(x = "marital_status", y = "age", hue = "adoption", col = "gender", palette = "Set2", data = marketing, kind = "swarm")
```

```python
sns.factorplot(x = "income", y = "education", hue = "adoption", col = "gender", palette = "Set1", data = marketing, kind = "swarm")
```
